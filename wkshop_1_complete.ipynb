1.4 Is there a relation between how many homeruns a player has made in a year and their salary that year? Pull Player's name, ID, Year, Salary and homeruns together in a single query.

```python
pd.read_sql("""
           SELECT 
               s.salary, s.yearID, s.playerID, b.HR, p.nameFirst, p.nameLast
           FROM
               salaries s
           INNER JOIN
               batting b
            ON
               s.playerID=b.playerID
            
            LEFT JOIN( 
               SELECT playerID, nameFirst, nameLast 
               FROM people) p
            ON
               s.playerID=p.playerID
               
            ORDER BY
               HR DESC
            LIMIT 30 """, conn)

#there is no relationship between homeruns and salary
```
